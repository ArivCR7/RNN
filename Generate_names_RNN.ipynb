```python
char_to_ix = {ch:i for i,ch in enumerate(chars)}
ix_to_char = {i:ch for i,ch in enumerate(chars)}
```

```python
#Define network parameters
vocab_length = len(chars)
hidden_size = 100
learning_rate = 1e-3
Wxh = np.random.randn(hidden_size,vocab_length) * 0.01
Whh = np.random.randn(hidden_size,hidden_size) * 0.01
bh = np.zeros((hidden_size,1)) 
Why = np.random.randn(vocab_length,hidden_size) * 0.01
by = np.zeros((vocab_length,1))

```

```python
def lossFun(X, Y, h):
    xs, hs, ys, ps = {}, {}, {}, {}
    hs[-1] = np.copy(h)
    loss=0
    
    for t in range(len(X)):
        xs[t] = np.zeros((vocab_length,1))
        xs[t][X[t]] = 1
        hs[t] = np.tanh(np.dot(Wxh, xs[t]) + np.dot(Whh,hs[t-1]) + bh)
        ys[t] = np.dot(Why, hs[t]) + by
        ps[t] = np.exp(ys[t])/np.sum(np.exp(ys[t]))
        loss += -np.log(ps[t][Y[t],0])
    dWxh, dWhh, dWhy, dbh, dby = np.zeros_like(Wxh), np.zeros_like(Whh), np.zeros_like(Why), np.zeros_like(bh), np.zeros_like(by)
    dhnext = np.zeros_like(hs[0])
    for t in reversed(range(len(X))):
        dy = np.copy(ps[t])
        dy[Y[t]] -= 1
        dWhy += np.dot(dy, hs[t].T)
        dby += dy
        dh = np.dot(Why.T, dy) + dhnext
        dhraw = (1 - hs[t] * hs[t]) * dh
        dbh += dhraw
        dWxh += np.dot(dhraw, xs[t].T)
        dWhh += np.dot(dhraw, hs[t-1].T)
        dhnext = np.dot(Whh.T, dhraw)
    for dparam in [dWxh, dWhh, dWhy, dbh, dby]:
        np.clip(dparam, -5, 5, out=dparam)
    
    return loss, dWxh, dWhh, dWhy, dbh, dby, hs[len(X)-1]
```

```python
def sample(hs):
        max_length = 10
        counter=0
        indices = []
        ix=-1
        xs = np.zeros((vocab_length,1))
        newline=char_to_ix['\n']
        while ( (counter <= max_length) and (ix != newline) ):     
            hs = np.tanh(np.dot(Wxh, xs) + np.dot(Whh,hs) + bh)
            ys = np.dot(Why, hs) + by
            ps = np.exp(ys)/np.sum(np.exp(ys))
            ix = np.random.choice(list(range(vocab_length)), p=ps.ravel())
            indices.append(ix)
            counter+=1
            xs = np.zeros((vocab_length,1))
            xs[ix] = 1
        if counter == 10:
            indices.append(char_to_ix['\n'])
        
        #print("Indeces: ",indices)
        txt=''.join([ix_to_char[l] for l in indices])
        #print("".join([ix_to_char[ix] for ix in indices]))
        print(str(txt))    
            
            
```

```python
n = 0
names = 7
mWxh, mWhh, mbh, mWhy, mby = np.zeros_like(Wxh), np.zeros_like(Whh), np.zeros_like(bh), np.zeros_like(Why), np.zeros_like(by)
smooth_loss = -np.log(1.0/vocab_length)*7
while n <= 100*1000:
    
    if n==0:
        hprev = np.zeros((hidden_size,1))
    with open("C:/Users/a.ezhilarasan/Documents/AI/DeepLearning/RNN/dinos.txt",'r',encoding='utf-8') as f:
        examples = f.readlines()
    examples = [x.lower().strip() for x in examples]
    np.random.shuffle(examples)
    index = n % len(examples)
    X = [None] + [char_to_ix[ch] for ch in examples[index]]
    Y = X[1:] + [char_to_ix['\n']]
    
    loss, dWxh, dWhh, dWhy, dbh, dby, hprev = lossFun(X, Y, hprev)
    smooth_loss = smooth_loss*0.999 + loss*0.001
    if n%1000 == 0:
        print("iter: ",n)
        print("smooth_loss: ",smooth_loss)
        print("***********Sample names**************")
        #seed=0
        for i in range(names):
            sample(hprev)
    for param, dparam, mem in zip([Wxh, Whh, Why, bh, by],
                                  [dWxh, dWhh, dWhy, dbh, dby],
                                  [mWxh, mWhh, mWhy, mbh, mby]):
        mem += dparam*dparam
        param += -learning_rate*(dparam/np.sqrt(mem + 1e-8))
    n+=1
```
